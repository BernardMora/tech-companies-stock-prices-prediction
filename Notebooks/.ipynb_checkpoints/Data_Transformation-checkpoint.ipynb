```python
import pandas as pd
from ta import add_all_ta_features
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

Define Data Transformation

```python
def transform_data(df):
    # Add technical indicators using 'ta' library (Feature Engineering)
    df = add_all_ta_features(df, open="open", high="high", low="low", close="close", volume="volume", fillna=True)

    # Normalize the data

    # Select columns to normalize (excluding non-numeric columns like 'ticker', 'is_extended_hours', etc.)
#     columns_to_normalize = ['open', 'high', 'low', 'close', 'volume']
    
    # Standardization (Z-score)
#     scaler_std = StandardScaler()
#     df[columns_to_normalize] = scaler_std.fit_transform(df[columns_to_normalize])

    # Min-Max Scaling
#     scaler_minmax = MinMaxScaler()
#     df[columns_to_normalize] = scaler_minmax.fit_transform(df[columns_to_normalize])

    # L2 Normalization (Euclidean Norm)
    close_prices_norm = np.linalg.norm(df['close'])
    print(close_prices_norm)
    df['close'] = np.array(df['close']) / close_prices_norm

    open_prices_norm = np.linalg.norm(df['open'])
    df['open'] = np.array(df['open']) / open_prices_norm

    high_prices_norm = np.linalg.norm(df['high'])
    df['high'] = np.array(df['high']) / high_prices_norm

    low_prices_norm = np.linalg.norm(df['low'])
    df['low'] = np.array(df['low']) / low_prices_norm

    volume_norm = np.linalg.norm(df['volume'])
    df['volume'] = np.array(df['volume']) / volume_norm


    # Look for significance on 'is_extended_hours' column
    for index, row in df.iterrows():
        if row['is_extended_hours'] == True:
            print("True!") # This won't ever print
            
    # Drop because there is no significance 
    df.drop('is_extended_hours', axis=1, inplace=True)
    
    df['date'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
    
    # Reversing the order of rows and changing the index to descending order
    """Retrieved from: 
    https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html
    https://stackoverflow.com/questions/20444087/right-way-to-reverse-a-pandas-dataframe
    """
    df = df[::-1].reset_index(drop=True)

    
    return df
```

```python
def graph(data, mensaje):
    plt.figure(figsize = (18,9))

    plt.plot(range(len(data['date'][:len(data['close'])])), data['close'][:len(data['close'])], label='Close Prices')

    plt.xticks(range(0,len(data['date']),500),data['date'][0::500],rotation=45)

    # (mensaje ejemplo) : Predicción de 2 redes LSTM con capa densa de salida y 20 épocas (55% train and 15% predict)
    plt.title(mensaje, fontsize=25)
    plt.xlabel('Date',fontsize=18)
    plt.ylabel('Normalized Close Prices',fontsize=18)

    plt.grid(axis = 'y', color = 'gray', linestyle = 'dashed')
    plt.legend()
    plt.show()
```
